# 2018 Representative to Congress for District 2 in Maine for Democratic Primary

```python
xl_cd2 = pandas.ExcelFile("voting_data/Maine/Maine/June 2018/Maine Democratic Primary for CD2 CVR.xlsx")
```

```python
xl_cd2.sheet_names  # see all sheet names
```

```python
df_cd2 = pandas.read_excel("voting_data/Maine/Maine/June 2018/Maine Democratic Primary for CD2 CVR.xlsx",
                       sheet_name=xl_cd2.sheet_names[0])
```

```python
df_cd2.shape
```

```python
df_cd2.columns
```

```python
df_cd2.head().T
```

```python
cd2_first_round_votes = df_cd2['DEM Rep. to Congress 1st Choice District 2 '].value_counts()
cd2_first_round_votes
```

```python
cd2_first_round_votes_per_candidate = cd2_first_round_votes[~cd2_first_round_votes.index.str.contains('vote', regex=False)]
cd2_first_round_votes_per_candidate
```

```python
cd2_first_round_votes_per_candidate.sum()
```

first round  percentage lead by each candidate

```python
(cd2_first_round_votes_per_candidate/cd2_first_round_votes_per_candidate.sum())*100
```
